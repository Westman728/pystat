### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

Object-oriented and pyplot/MATLAB style, object-oriented is the recommended because it can be more clear what is going on in the code

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

The figure is the entire picture, including axis, axes and other info.
An axe(s?) represents a subplot in a figure and contains the plotted data, axis ticks labels etc.
The artist objects are the visual elements we see on the screen and they are either Primitives(graphical objects) or Containers(where we put the primitives, like Axis, Axes or Figure)

### When plotting in Matplotlib, what is the expected input data type?

Answer: The expected input is generally Numpy Arrays, or objects that can be converted into Numpy Arrays, such as lists

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4,4)
y = x**2
```

```python
#"fig, axe" names are arbitrary conventional names, python unpacks the tuple created from plt.subplots() and assigns the first 
#element to fig, the second to axe.
fig, axe = plt.subplots()
axe.plot(x, y, color="red") #can make more changes to the visuals of line

axe.set_title("Plotting y=x^2")
axe.set_xlabel("x-axis")
axe.set_ylabel("y-axis")

#Object Oriented approach
```

```python
#assign any data, dont assign "fig, axe"
x = np.linspace(-4,4)
y = x**2

#create the line plot + labels
plt.plot(x, y)
plt.xlabel("x-axis")
plt.ylabel("y-ais")
plt.title("Pyplot approach")
#show graph
plt.show()


#Pyplot approach
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2
#plenty colors, markers etc available
#plt.scatter(random_data_x, random_data_y, color="firebrick", marker="*", s=5)


# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

#create subplots with 1 row, 2 columns
fig,ax = plt.subplots(1, 2, figsize = (15, 6))

#from understanding, 2d arrays are easier to call using ax[row, col] rather than indexing like this
#example; ax[0, 0] would refer to row 0, column 0, useful when handling many subplots
ax[0].scatter(random_data_x, random_data_y, color="lime", marker=".")
ax[0].set_title("Random data plot")
ax[0].set_xlabel("x-axis")
ax[0].set_ylabel("y-axis")
ax[0].set_facecolor("black")

ax[1].bar(names, values, color="firebrick")
ax[1].set_title("Fruit plot")
ax[1].set_xlabel("Fruit names")
ax[1].set_ylabel("Quantity")
ax[1].set_facecolor("darkgray")

plt.show()
```
